```python
def Ham1(n=7):
#     print(np.dot(Z(6,n),Z(7,n)).shape, np.dot(X(4,n),X(5,n)).shape)
          
    H = X(0,n) + X(1,n) + X(2,n)+ X(3,n) + X(4,n) + X(5,n) + X(6,n) +\
    4*np.dot(X(0,n),Z(2,n)) + 9*np.dot(Z(1,n),X(6,n)) + \
    9*np.dot(X(3,n),X(4,n)) + 13*np.dot(Z(5,n),Z(6,n))
    return H

def U(Ham,t,dt):
    return sLA.expm(-1j*Ham()*dt)

def get_purity(rho):
    purity = np.trace(np.dot(rho,rho))
    return purity

def F(rho1, rho2):
    return np.abs(np.trace(sLA.sqrtm(sLA.sqrtm(rho1) @ rho2 @ sLA.sqrtm(rho1))))**2

def unitary_evolution(rho0, Ham, dt,t0 = 0, t1 = 0.5):
    t_list = []
    U_list = []
    rho_list = []
    F_list = []

    rho = copy.copy(rho0)
    
    rho = rho0.copy()
    rho_list.append(rho)
    F_list.append(F(rho, rho))
    t_list.append(t0)

    t = t0
    


    while t <= t1:
        U_list.append(U(Ham,t,dt))
        t += dt
        print
        rho = np.dot(np.dot(U_list[-1],rho), U_list[-1].conj().T)
        t_list.append(t)
        rho_list.append(rho)
        F_list.append(F(rho, rho0))
        
    return t_list, F_list
```

```python
n = 7

np.random.seed(2)

state0 = Qpsi(1)
state1 = Qpsi(n-1)
state = Qpsi(n)
state0.build_random_state()
state1.build_random_state()

state.set_coefs(np.kron(state0.get_coefs(),state1.get_coefs()))
# state.build_random_state()
state.split_system(1,n-1)
state.get_coefs()
```

```python
rho0 = state.get_A_rho()
rho_init = state.get_rho()
rho0, get_purity(rho0)
```

```python
for dt in [0.01, 0.001]:
    t_unit, F_unit = unitary_evolution(rho_init, Ham1, dt = dt,t0 = 0, t1 = 0.5)
#     t_lindblad, F_lindblad = Lindblad_evolution(rho0, dt = dt,t0 = 0, t1 = 0.5)
    fig, ax = plt.subplots()
    ax.set_title(r'dt = {}'.format(dt), fontsize = 14)
    ax.plot(t_unit, F_unit, label = r'unitary evolution', linewidth = 3)
#     ax.plot(t_lindblad, F_lindblad, label = r'Lindblad evolution', linestyle = '-.')
    plt.grid()
    plt.xlabel(r', c', fontsize = 14)
    plt.ylabel(r' $F(\rho(0),\rho(t))$', fontsize = 14)
    plt.legend()
    plt.show()
```
